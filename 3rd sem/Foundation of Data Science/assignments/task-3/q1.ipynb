```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def hist(data, bins, title='Histogram', x_label='x', y_label='y'):
    plt.figure()
    plt.hist(data, bins=bins, linewidth=0.4, edgecolor="blue")
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
```

```python
roll = 34
mean_normal = roll
std_normal = roll/10
normal_data = np.random.normal(loc=mean_normal, scale=std_normal, size=1000)
hist(normal_data, 20, title='Normal Data', x_label='Values', y_label='Frequency')
```

```python
scale_exp = roll/10
exp = np.random.exponential(scale=scale_exp, size=1000)
hist(exp, 40, title='Exponential Data', x_label='Values', y_label='Frequency')
```

```python
poisson = np.random.poisson(lam=roll/10, size=10000)
hist(poisson, 15, title='Poisson Data', x_label='Values', y_label='Frequency')
```

```python
hypergeometric = np.random.hypergeometric(ngood=200, nbad=roll, nsample=50, size=1000)
hist(hypergeometric, 15, title='hypergeometric Data', x_label='Values', y_label='Frequency')
```

```python
binomial = np.random.binomial(n=20, p=0.9, size=1000)
hist(binomial, 20, title='Binomial Data', x_label='Values', y_label='Frequency')
```

```python
negative_binomial = np.random.negative_binomial(n=roll//2, p=roll/100, size=2000)
hist(negative_binomial, 30, title='Negative Binomial Data', x_label='Values', y_label='Frequency')
```
