2. Numpy assignment 2: Estimating value of Pi (using Monte-Carlo Simulation):
{
Hint:
x = a number between 1 and -1 generated by using uniform random number
generator (For example x = np.random.uniform(-1, 1))
y = a number between 1 and -1 generated by using uniform random number
generator
distance(x, y) = distance of a point(x, y) from origin
point_is_inside_unit_circle(x, y) is True if distance(x, y) is less than equal to 1

To estimate pi:
generate number of points (x1, y1) ... (xN, yN) using uniform random number
generator. Pi = (number_of_points_inside_unit_circle/ total_numbers_generated) * 4
}

```python
import numpy as np
```

```python
x = []
y = []
distance = []
point_is_inside_unit_circle = []
for i in range(500000):
    x.append(np.random.uniform(-1, 1))
    y.append(np.random.uniform(-1, 1))
    distance.append(np.sqrt(x[i]**2 + y[i]**2))
    point_is_inside_unit_circle.append(True if distance[i]<=1 else False)
```

```python
pi = (point_is_inside_unit_circle.count(True)/500000)*4
```
