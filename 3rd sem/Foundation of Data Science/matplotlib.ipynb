```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
ax.plot(x, y)
plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]
z = [0, 1, 8, 27, 64, 125]

fig = plt.figure()
ax = fig.subplots(subplot_kw={'projection': '3d'})
ax.set_xticks([0, 1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$', r'$\sum$'], fontsize=18)
ax.plot(x, y, z, color='green', lw='2',marker='*', markersize='10')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize = (20, 8))
for i in range(3):
    for j in range(3):
        x = [1, 2, 3, 4, 5]
        y = [3, 5, 4, 2, 4]
        axes[i][j].plot(x, y)
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
slope = 30
m = np.tan(slope * np.pi / 180)
y = m * x + 5
print(y)
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Sine Wave')
ax.set_xlabel('Range')
ax.text(3.2, 0, r'$\epsilon_i=1$')
ax.set_ylabel('Angle')
```

```python
data_fruit_sales_kg = {'apple': 10, 'orange': 25, 'mango': 50, 'lime':20}

fruit_name = list(data_fruit_sales_kg.keys())
fruit_sales = list(data_fruit_sales_kg.values())

fig, ax = plt.subplots()
ax.bar(x = fruit_name, height=fruit_sales, width=1, edgecolor = "white", linewidth=0.7)

ax.set_title("Fruit Sales")
ax.set_xlabel("Fruits")
ax.set_ylabel("Sales(kg)")
```

# 3D Surface Plot

```python
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig, ax = plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(x, y, z, cmap='viridis')
```

# Central Limit Theorem

```python
num_samples = 10000
np.random.seed(100)
```

```python
population = np.random.exponential(scale=2, size=num_samples)
```

```python
plt.hist(population, bins=200)
plt.title("Histogram of original distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
```
