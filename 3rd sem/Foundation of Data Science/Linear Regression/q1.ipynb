```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_simple_linear_regression(x, y_true, y_pred, title='Linear Regression', xlabel='x', ylabel='y'):
    plt.scatter(x, y_true)
    plt.plot(x, y_pred, 'g')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(linewidth= 0.4, alpha=0.7)
```

```python
advertisement = 2
base_sale = 5
random_scale = 5    
num_sample = 100

np.random.seed(40)

advertisement_tv = np.random.uniform(low=0, high=20, size=num_sample)
sale_true = base_sale + advertisement_tv + advertisement + np.random.uniform(low=random_scale, high=-random_scale, size=num_sample)
sale_pred = base_sale + advertisement * advertisement_tv

plot_simple_linear_regression(x= advertisement_tv, y_true = sale_true,y_pred=sale_pred, title='Sales vs Advertisement')
```
